```python
import numpy as np

def is_p6_color(p6_filename):
    with open(p6_filename, "rb") as f:
        # Read and check the header
        header = f.readline().decode("utf-8").strip()
        if header != "P6":
            return False

        # Skip comments if present
        while True:
            line = f.readline().decode("utf-8").strip()
            if not line.startswith("#"):
                break

        # Read width, height, and maxval
        width, height = map(int, line.split())
        maxval = int(f.readline())

        # Read binary pixel data
        pixel_data = np.frombuffer(f.read(), dtype=np.uint8)

    # Reshape the pixel data to get RGB values
    rgb_data = pixel_data.reshape((height, width, 3))

    # Check if the image is color (RGB)
    return rgb_data.shape[2] == 3

if __name__ == "__main__":
    p6_filename = "nagata.pgm"
    if is_p6_color(p6_filename):
        print("The P6 file contains color information (RGB).")
    else:
        print("The P6 file does not contain color information.")
```

```python
from PIL import Image
import numpy as np

def analyze_p6_colors(p6_filename):
    with open(p6_filename, "rb") as f:
        # Read and check the header
        header = f.readline().decode("utf-8").strip()
        assert header == "P6"

        # Skip comments if present
        while True:
            line = f.readline().decode("utf-8").strip()
            if not line.startswith("#"):
                break

        # Read width, height, and maxval
        width, height = map(int, line.split())
        maxval = int(f.readline())

        # Read binary pixel data
        pixel_data = np.frombuffer(f.read(), dtype=np.uint8)

    # Reshape the pixel data to get RGB values
    rgb_data = pixel_data.reshape((height, width, 3))

    # Convert to a set to get unique colors
    unique_colors = set(tuple(color) for row in rgb_data for color in row)

    return unique_colors

if __name__ == "__main__":
    p6_filename = "nagata.pgm"
    colors = analyze_p6_colors(p6_filename)

    print("Colors in the P6 PGM file:")
    for color in colors:
        print(color)
```

```python
import cv2
import numpy as np
from PIL import Image

def detect_edges(image_filename, output_filename):
    # Read the PGM file and convert to a NumPy array
    image = Image.open(image_filename)
    grayscale_image = np.array(image)

    # Apply Canny edge detector
    edges = cv2.Canny(grayscale_image, 30, 100)

    # Convert the edges to a PIL Image
    edges_image = Image.fromarray(edges)

    # Save the image with marked boundaries
    edges_image.save(output_filename)
    print(f"Edges marked image saved as {output_filename}")

if __name__ == "__main__":
    image_filename = "nagata.pgm"
    output_filename = "edges_marked_image.png"
    detect_edges(image_filename, output_filename)
```

```python
import cv2
import numpy as np
from PIL import Image, ImageDraw

def draw_edges(image_filename, output_filename, threshold=30):
    # Read the PGM file and convert to a NumPy array
    image = Image.open(image_filename)
    grayscale_image = np.array(image)

    # Calculate the gradient using Sobel
    gradient_x = cv2.Sobel(grayscale_image, cv2.CV_64F, 1, 0, ksize=3)
    gradient_y = cv2.Sobel(grayscale_image, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

    # Create a binary mask based on the threshold
    edge_mask = (gradient_magnitude > threshold).astype(np.uint8)

    # Convert the mask to a PIL Image
    edge_mask_image = Image.fromarray(edge_mask * 255)

    # Draw the edges on the original image
    result_image = Image.fromarray(grayscale_image)
    draw = ImageDraw.Draw(result_image)

    # Use any() to check if any element in the pixel position is equal to 1
    for y in range(edge_mask.shape[0]):
        for x in range(edge_mask.shape[1]):
            if edge_mask[y, x].any():
                draw.point((x, y), fill="black")

    # Save the image with marked edges
    result_image.save(output_filename)
    print(f"Edges marked image saved as {output_filename}")

if __name__ == "__main__":
    image_filename = "nagata.pgm"
    output_filename = "edges_marked_image.png"
    draw_edges(image_filename, output_filename)
```

```python
from PIL import Image, ImageDraw

def mark_origin(image_filename, output_filename, origin=(0, 0), side_length=10):
    # Open the existing image with marked edges
    image = Image.open(image_filename)

    # Calculate the center of the image
    center_x, center_y = image.size[0] // 2, image.size[1] // 2

    # Calculate the position of the specified origin relative to the center
    origin_x = center_x + origin[0]
    origin_y = center_y + origin[1]

    # Calculate the coordinates for the square
    x1 = max(0, int(origin_x - side_length / 2))
    y1 = max(0, int(origin_y - side_length / 2))
    x2 = min(image.size[0], int(origin_x + side_length / 2))
    y2 = min(image.size[1], int(origin_y + side_length / 2))

    # Draw a red square at the specified origin with the specified side length
    draw = ImageDraw.Draw(image)
    draw.rectangle([(x1, y1), (x2, y2)], fill="red")

    # Save the image with the red square at the origin
    image.save(output_filename)
    print(f"Red square at the origin added to the image and saved as {output_filename}")

if __name__ == "__main__":
    edges_image_filename = "edges_marked_image_1.png"
    marked_origin_output_filename = "marked_origin_image_square.png"
    #(-26.117105, -35.734393)
    custom_origin = (-522.3421, -714.68786)
    square_side_length = 20

    mark_origin(edges_image_filename, marked_origin_output_filename, origin=custom_origin, side_length=square_side_length)
```

```python
from PIL import Image, ImageDraw

def plot_rectangles(image_filename, output_filename, origins_and_points, rectangle_size=(20, 20), rectangle_color="orange"):
    # Open the existing image with marked edges
    image = Image.open(image_filename)

    # Calculate the center of the image
    center_x, center_y = image.size[0] // 2, image.size[1] // 2

    # Draw rectangles at the specified origins and points
    draw = ImageDraw.Draw(image)
    for origin, point in origins_and_points:
        # Calculate the position of the specified origin relative to the center
        origin_x = center_x + origin[0]
        origin_y = center_y + origin[1]

        # Calculate the coordinates for the rectangle
        x1 = max(0, int(origin_x + point[0] - rectangle_size[0] / 2))
        y1 = max(0, int(origin_y + point[1] - rectangle_size[1] / 2))
        x2 = min(image.size[0], int(origin_x + point[0] + rectangle_size[0] / 2))
        y2 = min(image.size[1], int(origin_y + point[1] + rectangle_size[1] / 2))

        # Draw an orange rectangle at the specified position
        draw.rectangle([(x1, y1), (x2, y2)], fill=rectangle_color)

    # Save the image with orange rectangles
    image.save(output_filename)
    print(f"Orange rectangles added to the image and saved as {output_filename}")

if __name__ == "__main__":
    edges_image_filename = "edges_marked_image_1.png"
    marked_rectangles_output_filename = "marked_rectangles_image.png"
    custom_origins_and_points = [
        ((-522.3421, -714.68786), (485.67503532, 706.58782385)),
        ((-522.3421, -714.68786), (463.02008163, 709.18757263)),
        ((-522.3421, -714.68786), (527.57625467, 712.84332856)),
        ((-522.3421, -714.68786), (1153.54020176, 664.82576146)),
        ((-522.3421, -714.68786), (648.20248058, 784.69559345)),
        ((-522.3421, -714.68786), (644.58765258, 699.33709907)),
        ((-522.3421, -714.68786), (868.76223122, 693.63108572)),
        ((-522.3421, -714.68786), (863.84097737, 777.50943446)),
    ]
    rectangle_size = (20, 20)

    plot_rectangles(edges_image_filename, marked_rectangles_output_filename, custom_origins_and_points, rectangle_size)
```
